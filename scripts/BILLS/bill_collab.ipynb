```python
import faiss
```

```python
index = faiss.read_index("bills.index")
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("nomic-ai/nomic-embed-text-v1.5", trust_remote_code=True)
```

```python
sentences = ['search_query: Has Congress done anything about its authority over tarrifs or taxes?']
query_embedding = model.encode(sentences)
```

```python
import numpy as np

# Correct the shape of query_embedding from (768, 1) to (1, 768)
# Assuming query_embedding is currently (768,)
query_embedding = np.array(query_embedding, dtype=np.float32).reshape(1, -1)

# Normalize the query embedding
normalized_query_embedding = query_embedding / np.linalg.norm(query_embedding)

# Perform the search
distances, indices = index.search(normalized_query_embedding, 5)

print("Distances:", distances)
print("Indices:", indices)
```
